### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here

df_filtered = df.dropna(subset = 'ConvertedCompYearly')

plt.figure(figsize=(10, 6))
plt.hist(df_filtered['ConvertedCompYearly'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Yearly Compensation')
plt.ylabel('Number of Respondents')
plt.title('Histogram of Total Compensation')
plt.tight_layout()
plt.show()

df_filtered = df_filtered.copy()
df_filtered['ConvertedCompYearly_log'] = np.log(df_filtered['ConvertedCompYearly'])
plt.figure(figsize=(10, 6))
plt.hist(df_filtered['ConvertedCompYearly_log'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Yearly Compensation (LOG transformed)')
plt.ylabel('Number of Respondents')
plt.title('Histogram of Total Compensation')
plt.tight_layout()
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here

df_age = df.dropna(subset = 'Age')
df_age = df_age[df_age['Age'] != 'Prefer not to say']

age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
  }

df_age['Age_numeric'] = df_age['Age'].map(age_map)

plt.figure(figsize=(8, 6))
plt.boxplot(df_age['Age_numeric'], tick_labels=['Age'])

plt.title("Box plot of age")
plt.ylabel('Age of respondents')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here
df_age_comp = df[['Age', 'ConvertedCompYearly']].dropna()

df_age_comp = df_age_comp[df_age_comp['Age'] != 'Prefer not to say']

age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
  }

df_age_comp['Age_numeric'] = df_age_comp['Age'].map(age_map)

df_age_comp = df_age_comp.copy()
df_age_comp['ConvertedCompYearly_log'] = np.log(df_age_comp['ConvertedCompYearly'])

plt.figure(figsize=(7, 6))
plt.scatter(df_age_comp['Age_numeric'], df_age_comp['ConvertedCompYearly'], alpha=0.5)

plt.xlabel('Age (approximate midpoint)')
plt.ylabel('Yearly compensation')
plt.title('Age vs. Yearly Compensation')
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(7, 6))
plt.scatter(df_age_comp['Age_numeric'], df_age_comp['ConvertedCompYearly_log'], alpha=0.5)

plt.xlabel('Age (approximate midpoint)')
plt.ylabel('Yearly compensation (log transformed)')
plt.title('Age vs. Yearly Compensation')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here

df_sat_comp_age = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Age']].dropna()

df_sat_comp_age = df_sat_comp_age[df_sat_comp_age['Age'] != 'Prefer not to say']

age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
  }

df_sat_comp_age['Age_numeric'] = df_sat_comp_age['Age'].map(age_map)

plt.figure(figsize=(10, 7))
plt.scatter(df_sat_comp_age['ConvertedCompYearly'], df_sat_comp_age['JobSatPoints_6'], s=df_sat_comp_age['Age_numeric'] * 5, alpha=0.3)

plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction')
plt.title('Bubble Plot: Compensation vs. Job Satisfaction')
plt.grid(True)
plt.show()

df_sat_comp_age = df_sat_comp_age.copy()
df_sat_comp_age['ConvertedCompYearly_log'] = np.log(df_sat_comp_age['ConvertedCompYearly'])

plt.figure(figsize=(10, 7))
plt.scatter(df_sat_comp_age['ConvertedCompYearly_log'], df_sat_comp_age['JobSatPoints_6'], s=df_sat_comp_age['Age_numeric'] * 5, alpha=0.3)

plt.xlabel('Yearly Compensation (log transformed)')
plt.ylabel('Job Satisfaction')
plt.title('Bubble Plot: Compensation vs. Job Satisfaction')
plt.grid(True)
plt.show()
```
